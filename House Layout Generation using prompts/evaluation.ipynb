```python
def calculate_iou(bbox1, bbox2):
    """
    Calculate Intersection over Union (IoU) between two bounding boxes.
    :param bbox1: Bounding box coordinates [x_min, y_min, x_max, y_max].
    :param bbox2: Bounding box coordinates [x_min, y_min, x_max, y_max].
    :return: Intersection over Union (IoU) value.
    """
    x_min1, y_min1, x_max1, y_max1 = bbox1
    x_min2, y_min2, x_max2, y_max2 = bbox2
    
    # Calculate intersection area
    x_left = max(x_min1, x_min2)
    y_top = max(y_min1, y_min2)
    x_right = min(x_max1, x_max2)
    y_bottom = min(y_max1, y_max2)

    intersection_area = max(0, x_right - x_left + 1) * max(0, y_bottom - y_top + 1)

    # Calculate union area
    bbox1_area = (x_max1 - x_min1 + 1) * (y_max1 - y_min1 + 1)
    bbox2_area = (x_max2 - x_min2 + 1) * (y_max2 - y_min2 + 1)
    union_area = bbox1_area + bbox2_area - intersection_area

    # Calculate IoU
    iou = intersection_area / union_area
    return iou

# Given bounding boxes data
bounding_boxes = [
    {"bbox": [35.0, 50.0, 65.0, 80.0], "color": "#38e1f3", "name": "Living Room", "direction": "North"},
    {"bbox": [30.0, 30.0, 50.0, 50.0], "color": "#d24758", "name": "Bedroom", "direction": "South west"},
    {"bbox": [50.0, 50.0, 65.0, 70.0], "color": "#397b7e", "name": "Kitchen", "direction": "North east"},
    {"bbox": [50.0, 35.0, 80.0, 50.0], "color": "#884b4e", "name": "Bedroom 2", "direction": "South east"}
]

# Calculate IoU for each pair of bounding boxes
iou_values = []
for i in range(len(bounding_boxes)):
    for j in range(i + 1, len(bounding_boxes)):
        bbox1 = bounding_boxes[i]["bbox"]
        bbox2 = bounding_boxes[j]["bbox"]
        iou = calculate_iou(bbox1, bbox2)
        iou_values.append(iou)

# Calculate mean IoU
mean_iou = sum(iou_values) / len(iou_values)

# Print results
print("Intersection over Union (IoU) values:")
for i, iou in enumerate(iou_values, start=1):
    print(f"IoU {i}: {iou:.4f}")
print(f"Mean IoU: {mean_iou:.4f}")
```
