```python
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import PeftModel, PeftConfig
```

```python
base_model = "meta-llama/Meta-Llama-3-8B"
adapter_model = "kishorep3105/metallama-house-v3"
```

```python
model = AutoModelForCausalLM.from_pretrained(base_model,device_map="auto",
    torch_dtype='auto')
```

```python
model.device
```

```python
model = PeftModel.from_pretrained(model, adapter_model,device_map="auto",
    torch_dtype='auto')
```

```python
model.device
```

```python
tokenizer = AutoTokenizer.from_pretrained(base_model)
```

```python
import torch
# monkey patched for quanto
def named_module_tensors(module, recurse=False):
    for named_parameter in module.named_parameters(recurse=recurse):
      name, val = named_parameter
      flag = True
      if hasattr(val,"_data") or hasattr(val,"_scale"):
        if hasattr(val,"_data"):
          yield name + "._data", val._data
        if hasattr(val,"_scale"):
          yield name + "._scale", val._scale
      else:
        yield named_parameter

    for named_buffer in module.named_buffers(recurse=recurse):
      yield named_buffer

def dtype_byte_size(dtype):
    """
    Returns the size (in bytes) occupied by one parameter of type `dtype`.
    """
    import re
    if dtype == torch.bool:
        return 1 / 8
    bit_search = re.search(r"[^\d](\d+)$", str(dtype))
    if bit_search is None:
        raise ValueError(f"`dtype` is not a valid dtype: {dtype}.")
    bit_size = int(bit_search.groups()[0])
    return bit_size // 8

def compute_module_sizes(model):
    """
    Compute the size of each submodule of a given model.
    """
    from collections import defaultdict
    module_sizes = defaultdict(int)
    for name, tensor in named_module_tensors(model, recurse=True):
      size = tensor.numel() * dtype_byte_size(tensor.dtype)
      name_parts = name.split(".")
      for idx in range(len(name_parts) + 1):
        module_sizes[".".join(name_parts[:idx])] += size

    return module_sizes
```

```python
module_sizes = compute_module_sizes(model)
#print(module_sizes)
print(f"The model size is {module_sizes[''] * 1e-9} GB")
```

```python
text="""You are an AI assistant specializing in generating house layouts in HTML based on textual descriptions. Your task is to carefully read the provided instructions, understand the requirements for each room (category, size, position, adjacencies), and generate an SVG layout adhering to those specifications. Generate your SVG layout here, using <rect> elements with the following attributes: data-category: Room category (e.g. living room, kitchen, bathroom); x, y: Position coordinates; width, height: Size dimensions. Ensure rooms are positioned and sized according to the instructions.\n\n### Instruction:\nMake a common room 1 . The common room 1 should be at the south side of the apartment. It would be great to have common room 1 approx 200 sqft with the aspect ratio of 9 over 14. The common room 1 should have an en-suite bathroom. The common room 1 connects to the storage, master room, common room 2, living room. Can we have a storage ? The storage should be at the south west corner of the apartment. The storage should be approx 10 sqft with the aspect ratio of 14 over 11. The storage should be next to the common room 1, master room, living room. Can you make a master room ? Place master room at the south side of the apartment. Make master room about 200 sqft with the aspect ratio of 9 over 14. The master room should have an en-suite bathroom. The master room attaches to the common room 1, storage, common room 2, living room. Make a common room 2 . The common room 2 should be at the north side of the apartment. Can you make common room 2 around 100 sqft with the aspect ratio of 3 over 5? The common room 2 should have an en-suite bathroom. The common room 2 should be next to the common room 1, master room, living room, kitchen, bathroom. Can you make a living room ? The living room should be about 250 sqft with the aspect ratio of 1 over 2. Can you make a kitchen ? I would like to place kitchen at the north side of the apartment. Can we have kitchen to be approx 150 sqft with the aspect ratio of 9 over 14? The kitchen connects to the common room 2. The bathroom should be considered. Place bathroom at the north side of the apartment. Can we have bathroom to be approx 50 sqft with the aspect ratio of 3 over 8? The bathroom is in an en-suite bathroom. The bathroom connects to the common room 2. \n### Response:"""
```
